```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style='darkgrid')

grocery = pd.read_csv("Groceries_dataset.csv", parse_dates= ['Date'])
print(grocery.shape)
grocery.head()
```

```python
marketing = pd.read_csv("DirectMarketing.csv")
print(marketing.shape)
marketing.head()
```

```python
items = grocery[['Date','itemDescription']].groupby('Date').count().reset_index()
items.rename(columns={'itemDescription':'itemCount'}, inplace=True)
items.head()
```

```python
sns.relplot(x = 'Date', y = 'itemCount', data = items[items.Date > '2015-06-01'], 
            kind = 'line', height = 5, aspect = 2)
```
