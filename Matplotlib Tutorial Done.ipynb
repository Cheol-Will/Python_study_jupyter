## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = list(range(6))
y = list(range(0, 12, 2))

# Resize your Graph
plt.figure(figsize = (12, 8))
           
# plt.plot(x, y, label = "11x" , C = "red", linewidth = 2, marker = ".", linestyle = "--",  markersize = 10, markeredgecolor = "blue")

# Use shorthand notation
# fmt = "[color][marker][line]"
plt.plot(x, y, "b^--", label = "2x")


## Line Number Two
x2 = np.arange(0, 4, 0.5)
plt.plot(x2[:5], x2[:5]**2, "r", label = "x**2")
plt.plot(x2[4:], x2[4:]**2, "r--")


plt.title("Our First Graph!", fontdict = {"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("X Axis!", fontdict = {"fontname": "Arial"})
plt.ylabel("Y Axis!", fontdict = {"fontname": "Comic Sans MS"})

plt.xticks([0, 1, 2, 3, 4, 5])
plt.yticks(list(range(0, 12, 2)))

plt.legend()

plt.savefig("MyGraph.png")

plt.show()
```
