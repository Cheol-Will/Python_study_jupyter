```python
india=corona_virus[corona_virus['Country/Region']=='India'].groupby('ObservationDate').sum().reset_index()
us = corona_virus[corona_virus['Country/Region']=='US'].groupby('ObservationDate')['Confirmed'].sum().reset_index()
brazil = corona_virus[corona_virus['Country/Region']=='Brazil'].groupby('ObservationDate')['Confirmed'].sum().reset_index()
```

```python
fig = plt.figure(figsize=(20,12))
plt.stackplot(temp['ObservationDate'][:300], us['Confirmed'][:300],india['Confirmed'][:300],brazil['Confirmed'][:300], labels=['US', 'India', 'Brazil'])
plt.xticks(rotation=90)
plt.legend(loc='upper left',fontsize=20)

plt.show()
```

# Alignment

### subplot의 크기가 같을 때

```python
fig, axis = plt.subplots(2, 3, figsize = (8,5))
plt.tight_layout()  # subplot사이 간격 변경
plt.show()
```

### subplot의 크기가 다를 때

```python
fig = plt.figure(figsize=(8, 5))

gs = fig.add_gridspec(2, 2) # make 2 by 2 grid (row, col)

ax = [None for i in range(5)]

ax[0] = fig.add_subplot(gs[0, :]) 
ax[0].set_title('gs[0, :]')

ax[1] = fig.add_subplot(gs[1, :-1])
ax[1].set_title('gs[1, :-1]')

ax[2] = fig.add_subplot(gs[1:, -1])
ax[2].set_title('gs[1:, -1]')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 5))

gs = fig.add_gridspec(3, 3) # make 3 by 3 grid (row, col)

ax = [None for _ in range(5)]

ax[0] = fig.add_subplot(gs[0, :]) 
ax[0].set_title('gs[0, :]')

ax[1] = fig.add_subplot(gs[1, :-1])
ax[1].set_title('gs[1, :-1]')

ax[2] = fig.add_subplot(gs[1:, -1])
ax[2].set_title('gs[1:, -1]')

ax[3] = fig.add_subplot(gs[-1, 0])
ax[3].set_title('gs[-1, 0]')

ax[4] = fig.add_subplot(gs[-1, -2])
ax[4].set_title('gs[-1, -2]')

for ix in range(5):
    ax[ix].set_xticks([])
    ax[ix].set_yticks([])

plt.tight_layout()
plt.show()
```

# 예시

```python
out1=diabetes[diabetes.Outcome==1]
out0=diabetes[diabetes.Outcome==0]
```

```python
fig = plt.figure(figsize=(12, 8))

gs = fig.add_gridspec(2, 2) # make 3 by 3 grid (row, col)

ax = [None for _ in range(3)]

ax[0] = fig.add_subplot(gs[0, :]) 
ax[0].set_title('Histogram of Age')
ax[0].hist(diabetes['Age'], color='green')

ax[1] = fig.add_subplot(gs[1:,0])
ax[1].set_title('OutCome : 1')
ax[1].boxplot([out1['Age'], out1['BMI'], out1['BloodPressure'], out1['Glucose']],vert=False)
ax[1].set_xticks(ticks=np.arange(0,225, 25))
ax[1].set_yticks(ticks=[1,2,3,4])
ax[1].set_yticklabels(['Age', 'BMI', 'BloodPressure','Glucose'])

ax[2] = fig.add_subplot(gs[1:,1])
ax[2].set_title('OutCome : 0')
ax[2].boxplot([out0['Age'], out0['BMI'], out0['BloodPressure'], out0['Glucose']],vert=False)
ax[2].set_xticks(ticks=np.arange(0,225, 25))
ax[2].set_yticks(ticks=[1,2,3,4])
ax[2].set_yticklabels(['Age', 'BMI', 'BloodPressure','Glucose'])


plt.tight_layout()
plt.show()
```
