```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
covid_19 = pd.read_csv("./DScover_2nd_seminar_data_set/covid_19_data.csv", parse_dates=['ObservationDate'])
# parse_datse는 날짜를 분석하기 위함parse_datse는 날짜를 분석하기 위함
latest_covid_cases = pd.read_csv("./DScover_2nd_seminar_data_set/cases_country.csv")
diabetes = pd.read_csv("./DScover_2nd_seminar_data_set/diabetes.csv")
heart = pd.read_csv("./DScover_2nd_seminar_data_set/heart.csv")
heart_failure_clinical = pd.read_csv("./DScover_2nd_seminar_data_set/heart_failure_clinical_records_dataset.csv")
iris = pd.read_csv("./DScover_2nd_seminar_data_set/Iris.csv")
pokemon = pd.read_csv("./DScover_2nd_seminar_data_set/Pokemon.csv")
titanic = pd.read_csv("./DScover_2nd_seminar_data_set/train.csv")
```

## 1차원 실수 분포 plot: rug plot, kernel density, distplot, countplot

```python
x = iris["PetalLengthCm"].values

sns.rugplot(x)
plt.title("Iris PetalLengthCm Rug Plot")
plt.show()
```

```python
sns.kdeplot(x)
plt.title("Iris PetalLengthCm Kernel Density Plot")
plt.show()
```

```python
sns.distplot(x, kde = True, rug = True)
plt.title("Iris PetalLengthCm Dist Plot")
plt.show()
```

#### countplot for only dataframe

```python
sns.countplot(x = "Pclass", data = titanic)
plt.title("Per class, number of people")
plt.show()
```

## 2차원 실수 분포 plot: jointplot, pairplot, heatmap

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", data = iris)
plt.suptitle("SepalLengthCm & SepalWidthCm Joint Plot", y = 1.02)
plt.show()
```

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", data = iris, kind = "kde")
plt.suptitle("SepalLengthCm & SepalWidthCm Joint Plot", y = 1.02)
plt.show()
```

```python
sns.pairplot(iris.drop(["Id"], axis = 1), hue = "Species", markers = ["o", "s", "D"])
plt.title("Iris Species Pair Plot, using Hue")
plt.show()
```

```python
titanic_size = titanic.pivot_table(index = "Pclass", columns = "Sex", aggfunc = "size")
titanic_size
```

```python
titanic.groupby(["Pclass", "Sex"]).count()
```

```python
sns.heatmap(titanic_size, cmap = sns.light_palette("gray", as_cmap = True), annot = True, fmt = "d")
plt.title("Heatmap")
plt.show()
```

#### 2-dimentional data (category & number)

```python
sns.barplot(x = "day", y = "total_bill", data = tips) # 평군과, 편차
plt.title("Per Day Total Tip Bar Plot")
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips)
plt.title("Per Day Total Tip Box Plot")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips) # 세로 방향으로 kde density histogram
plt.title("Per Day Total Tip Violin Plot")
plt.show()
```

```python
np.random.seed(0)
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True) # scatter와 유사
plt.title("Per Day Total Tip Strip Plot") # jitter로 안 겹치게 범주별 x값 약간씩 조정
plt.show()
```

```python
sns.swarmplot(x = "day", y = "total_bill", data = tips) # scatter와 유사
plt.title("Per Day Total Tip Swarm Plot") # 아예 안 겹치게 범주별 x값 조정
plt.show()
```
