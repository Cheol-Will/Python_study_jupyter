```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1 = np.random.rand(100)
x2 = np.random.rand(100)
w1 = np.random.uniform(low=-1.0, high=1.0)
w2 = np.random.uniform(low=-1.0, high=1.0)
b = np.random.uniform(low=-1.0, high=1.0)
```

```python
x1 = X1
a1 = w1
a2 = w2
rb = b
```

```python
rng = 2
x1 = x1 * rng
```

```python
y = 0.3 * x1 + 0.5 * x2 + rb

errors = []
w1_grad = []
w2_grad = []

num_epoch=1000
learning_rate= 0.1
```

```python
for epoch in range(num_epoch):
    # 예측값
    y_hat = a1 * x1 + a2 * x2 + rb

    error = ((y_hat - y)**2).mean()
    if error < 0.00001:
        break
    
    # 미분값 적용 (Gradient)
    a1 = a1 - learning_rate * ((y_hat - y) * x1).mean()
    a2 = a2 - learning_rate * ((y_hat - y) * x2).mean()
    
    w1_grad.append(a1)
    w2_grad.append(a2)
    
    rb = rb - learning_rate * (y_hat - y).mean()
    
    errors.append(error)
    
    if epoch % 5 == 0:
        print("{0:2} a1 = {1:.5f}, a2 = {2:.5f}, rb = {3:.5f} error = {4:.5f}".format(epoch, a1, a2, rb, error))

print("----" * 15)
print("{0:2} a1 = {1:.1f}, a2 = {2:.1f}, rb = {3:.1f} error = {4:.5f}".format(epoch, a1, a2, rb, error))
```

```python
plt.figure(figsize=(10, 7))

plt.plot(errors)
```

```python
plt.figure(figsize=(10, 7))

plt.hlines(y=0.3, xmin=0, xmax=len(w1_grad), color='r')
plt.plot(w1_grad, color='g')
plt.ylim(0, 1)
plt.title('W1', fontsize=16)
plt.legend(['W1 Change', 'W1'])
plt.show()
```

```python
plt.figure(figsize=(10, 7))

plt.hlines(y=0.5, xmin=0, xmax=len(w2_grad), color='r')
plt.plot(w2_grad, color='g')
plt.ylim(0, 1)
plt.title('W2', fontsize=16)
plt.legend(['W2 Change', 'W2'])
plt.show()
```
